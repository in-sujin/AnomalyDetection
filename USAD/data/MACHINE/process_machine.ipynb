<b>train

```python
train = pd.read_csv('11_lcahu_feature_train.csv', header=0, low_memory=False)
```

```python
train.head()
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()

x = train.values
x_scaled = min_max_scaler.fit_transform(x)
norm_train = pd.DataFrame(x_scaled)
```

```python
norm_train.head()
```

```python
norm_train.to_csv('train.csv', index=False)
```

```python
test = pd.read_csv('11_lcahu_feature_test.csv', header=0, low_memory=False)
```

```python
test.head()
```

```python
labels = pd.DataFrame(test["attack"])
test = test.drop(columns='attack')
```

```python
norm_test.head()
```

```python
norm_test.to_csv('test.csv', index=False)
```
