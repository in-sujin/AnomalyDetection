```python
import pandas as pd
```

---

<b>train

```python
train = pd.read_csv('SWaT_Dataset_Normal_v0.csv', header=1, low_memory=False)
```

```python
train.head()
```

```python
train = train.drop(columns=[' Timestamp', 'attack'])
```

```python
train.head()
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()

x = train.values
x_scaled = min_max_scaler.fit_transform(x)
norm_train = pd.DataFrame(x_scaled)
```

```python
norm_train.head()
```

```python
norm_train.to_csv('train.csv', index=False)
```

<b> test

```python
test = pd.read_csv('SWaT_Dataset_Attack_v0.csv', header=0, low_memory=False)
```

```python
test.head()
```

```python
labels = [ float(label!= 'Normal' ) for label  in test["attack"].values]
test = test.drop(columns=[' Timestamp', 'attack'])
```

```python
labels = pd.DataFrame(labels)
```

```python
labels
```

```python
test.head()
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()

x = test.values
x_scaled = min_max_scaler.fit_transform(x)
norm_test = pd.DataFrame(x_scaled)
```

```python
norm_test.head()
```

```python
norm_test.to_csv('test.csv', index=False)
```

```python
labels.to_csv('labels.csv', index=False)
```
